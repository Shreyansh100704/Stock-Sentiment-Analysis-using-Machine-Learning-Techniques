```python
import requests
import json
import csv
from datetime import datetime
import pandas as pd
import time
```

```python
# Function to get articles for a specific month
def get_articles(api_key, year, month, max_retries=3):
    url = f"https://api.nytimes.com/svc/archive/v1/{year}/{month}.json"
    params = {
        'api-key': api_key
    }

    response = requests.get(url, params=params)
    if response.status_code == 200:
        print(f"Successfully retrieved articles for {year}-{month}")
        time.sleep(12)
        return response.json()

    elif response.status_code == 403:
      pass

    else:
        print(f"Failed to retrieve articles for {year}-{month}: {response.status_code}")
        time.sleep(12)
        return None


# Function to fetch headlines and dates for the past 10 years
def fetch_headlines(api_key, years=10):
    current_date = datetime.now()
    headlines = []

    for year_offset in range(years):
        year = current_date.year - year_offset

        if year_offset == 0:
          for month in range(1, 7):
            try:
                print(f"Fetching articles for {year}-{month:02d}")
                data = get_articles(api_key, year, month)
                for doc in data['response']['docs']:
                    headline = doc.get('headline', {}).get('main', '')
                    pub_date = doc.get('pub_date', '')
                    headlines.append({
                            'headline': headline,
                            'pub_date': pub_date
                        })

            except requests.exceptions.RequestException as e:
                print(e)
                continue

            except (KeyError, json.JSONDecodeError) as e:
                print(f"Error parsing data for {year}-{month:02d}: {e}")
                continue

        else:
          for month in range(1, 13):
              try:
                  print(f"Fetching articles for {year}-{month:02d}")
                  data = get_articles(api_key, year, month)
                  for doc in data['response']['docs']:
                      headline = doc.get('headline', {}).get('main', '')
                      pub_date = doc.get('pub_date', '')
                      headlines.append({
                              'headline': headline,
                              'pub_date': pub_date
                          })

              except requests.exceptions.RequestException as e:
                  print(e)
                  continue

              except (KeyError, json.JSONDecodeError) as e:
                  print(f"Error parsing data for {year}-{month:02d}: {e}")
                  continue

    return headlines



# Save headlines to a CSV file
def save_to_csv(data, filename='nyt_headlines_2015_to_2024.csv'):
    with open(filename, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['Headline', 'Publication Date'])
        for item in data:
            writer.writerow([item['headline'], item['pub_date']])
```
